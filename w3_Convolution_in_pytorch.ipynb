Convolution in Pytorch

```python
import torch
import torchvision

import matplotlib.pyplot as plt
import numpy as np
```

```python
imt = torchvision.io.read_image('CG - 500.PNG')
imt = imt/255.0

imt = imt.unsqueeze(0)
print(np.ones((imt.shape[1], 5, 5)))
```

```python
with torch.inference_mode():
    w = np.ones((imt.shape[1], 5, 5)) / (imt.shape[1]*5*5)
    
    w = w.astype(np.float32)
    
    weight = torch.from_numpy(w).unsqueeze(0)
    
    result = torch.conv2d(imt, weight)
    
    result = result.squeeze(0).permute(1, 2, 0)
    
plt.imshow(result.numpy(), cmap='gray')
plt.show()
```
